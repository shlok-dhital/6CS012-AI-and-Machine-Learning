# **3 Image Compression and Decompression using PCA.**

## 3.1 Load and Prepare Data:

```python
# Load image in greyscale
img = Image.open("/content/drive/MyDrive/AI - ShlokDhital/chaewon2.png").convert('L')
img_arr = np.array(img)
print("Shape:", img_arr.shape)
plt.imshow(img_arr, cmap='gray')
plt.axis('off')
plt.show()

# Standardize the data by subtracting mean
height, width = img_arr.shape
data = img_arr.reshape(height,width)

mean = np.mean(data,axis=0)
centered_data = data - mean

# Calculate covariance matrix
cov_matrix = np.cov(centered_data, rowvar=False)
```

## 2. Eigen Decomposition and Identifying Principal Components

```python
# Compute Eigenstuff
eigenvalues , eigenvectors = np.linalg.eigh(cov_matrix)

# Sort eigenvectors and eigenvalues in descending order
sorted_idx = np.argsort(eigenvalues)[::-1]
eigenvalues = eigenvalues[sorted_idx]
eigenvectors = eigenvectors[:,sorted_idx]

# Plot the explained variance ratio
explained_variance_ratio = eigenvalues / np.sum(eigenvalues)
cum_sum = np.cumsum(explained_variance_ratio)
plt.plot(cum_sum)
plt.vlines(100, cum_sum.min(), 1, colors="red", linestyles="dashed", label="k = 100")
plt.title("Cumulative Explained Variance")
plt.xlabel("Number of Components")
plt.ylabel("Cumulative Explained Variance")
plt.legend()
plt.grid(True)
plt.show()
```

## Reconstruction and Experiment

```python
# k = 100 covers almost all data
k = 100
components = eigenvectors[:,:k]

compressed_data = np.dot(centered_data, components)

decompressed_data = np.dot(compressed_data, components.T) + mean

print(decompressed_data.shape)

plt.figure(figsize=(12,3))

# Original
plt.subplot(1,3,1)
plt.imshow(img_arr, cmap = "gray")
plt.title(f"Original {img_arr.shape}")
plt.axis('off')

# Compressed
plt.subplot(1,3,2)
plt.imshow(compressed_data, cmap = "gray")
plt.title(f"Compressed {compressed_data.shape}")
plt.axis('off')

# Decompressed
plt.subplot(1,3,3)
plt.imshow(decompressed_data, cmap = "gray")
plt.title(f"Decompressed {decompressed_data.shape}")
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
k_values = [10, 100, 1000, 1500]
no_k_values = len(k_values)
height_per_row = 3

fig, axes = plt.subplots(nrows=no_k_values, ncols=3,
                         figsize=(16, no_k_values * height_per_row))

for i, k in enumerate(k_values):
    components = eigenvectors[:, :k]
    compressed_data = np.dot(centered_data, components)
    decompressed_data = np.dot(compressed_data, components.T) + mean

    row_axes = axes[i]


    # Compressed
    row_axes[0].imshow(compressed_data, cmap="gray")
    row_axes[0].set_title(f"Compressed {compressed_data.shape}")
    row_axes[0].axis('off')

    # Original
    row_axes[1].imshow(img_arr, cmap="gray")
    row_axes[1].set_title(f"Original {img_arr.shape}")
    row_axes[1].axis('off')

    # Decompressed
    row_axes[2].imshow(decompressed_data, cmap="gray")
    row_axes[2].set_title(f"Decompressed {decompressed_data.shape}")
    row_axes[2].axis('off')

plt.tight_layout()
plt.show()
```

### Evaluation
From the plot above that shows compression using 10, 100, 1000 and 1500 principle components, we can see that the quality of decompressed image increases as the number of principle components is increased. The change is most evident in the jump from 10 to 100 and becomes less prominent henceforth. The jump from 100 to 1000 is not as significant as from 10 to 100, as predicted by our cumulative sum plot
